## Cleaning
Let's suppose (**as it should always be**) that we are only interested in rows where the p-values values and the adjusted p-values are < 0.05 

```python
# Filtering data
df_time1f = df_time1[ (df_time1['edger.p.value'] <= 0.05) & (df_time1['edger.adj.p.value'] <= 0.05) ]
df_time1f
```

```python
# I use the function list to oblige pandas to print the whole line without '...'
list(df_time1f.loc[5,['sprot_Top_BLASTX_hit']])
```

```python
# Filter lines where the organisms of the BLAST best hits are plants
df_time1f_plant = df_time1f[ df_time1f['sprot_Top_BLASTX_hit'].str.contains("Viridiplantae")]

df_time1f_plant
```
